## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

First we have to apply all the conditions to generate the total revenue of the E-Commerece company. Then using sum() function we will get the total revenue

```python
df1 = pd.DataFrame(ECom_Data.groupby(['Brand']).Customer_uniq_id.count())
df2 = pd.DataFrame(ECom_Data.groupby(['Brand']).discounted_price.sum())
df=pd.concat([df1,df2],axis=1)
df.columns=['order_no','brand_revenue']

def revenue(x,y):
    if x>600:
        return y*0.25
    elif x>350 and x<=600:
        return y*0.15
    elif x>100 and x<=350:
        return y*0.10
    else:
        return y*0.05
    
   
```

```python
 df['revenue']=df.apply(lambda x: revenue(x['order_no'], x['brand_revenue']),axis=1)
```

In below cell we use round() function to round up the output (to avoid the float value).

```python
round(df['revenue'].sum())
```

The total Revenue of E-Commerece company is 611061.

## Q 8. Create a new dataframe to include the Brand specific information as stated:
## i.	total number of orders placed per Brand
## ii.	total retail_price per Brand
## iii.	total discounted_price per Brand, and 
## iv.	total BrandRevenue generated per Brand. 

## Also, draw a pairplot using these four features. (6 marks)


### Ans 8

```python
# df1 = Total number of orders placed per Brand 
df1=pd.DataFrame(ECom_Data.groupby(['Brand']).Customer_uniq_id.count())

# df2 = Total retail_price per Brand
df2=pd.DataFrame(ECom_Data.groupby(['Brand']).retail_price.sum())

# df3 = Total discounted_price per Brand
df3=pd.DataFrame(ECom_Data.groupby(['Brand']).discounted_price.sum())

# df4 = Total Brand Revenue generated per Brand
df4=pd.DataFrame(ECom_Data.groupby(['Brand']).retail_price.sum()-ECom_Data.groupby(['Brand']).discounted_price.sum())
```

```python
# Creating new data frame
new_df=pd.concat([df1,df2,df3,df4],axis=1)
new_df.columns=['Order Count','Retail Price','Discounted Price','Brand Revenue']
new_df
```

```python
# to avoid warnings we import filterwarnings. 
import warnings
warnings.filterwarnings("ignore")
```

```python
def rev_for_each(x):
    if x>600:
        return x*0.25
    elif x>350 and x<=600:
        return x*0.15
    elif x>100 and x<=350:
        return x*0.10
    else:
        return x*0.05
  
```
